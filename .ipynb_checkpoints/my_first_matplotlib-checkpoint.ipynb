```python
from matplotlib import pyplot as plt
```

```python
plt.title("My title")
plt.xlabel("x")
plt.ylabel("y")
plt.plot([1,2,3], [4,5,6], label="line", color='red')
plt.scatter([2.25, 1.75, 2], [5.75, 5.25, 5.5], label="scatter", color='blue')
plt.bar([2.5, 2.75, 3], [4, 4.5, 4.25], label="bar", color='#666666')
plt.legend()
plt.show()
```
